```python
import numpy as np
import matplotlib.pyplot as plt
```

## Проверяем условие несмещенности 

```python
n=10
M=10000
X =np.random.normal(0,1, (M, n))
V = X.mean(axis = 1)
print(V.mean())
plt.scatter(range(M), V)
```

```python
plt.plot(V)
```

## Проверяем условие состоятельности

```python
V = []
for n in range(1, M + 1):
    x = np.random.normal(0, 1, n)
    V.append(x.mean())
plt.plot(V)
```

## Проверяем асимптотическую нормальность

```python
n=10
M=100
X =np.random.normal(0,1, (M, n))
V = X.mean(axis = 1)
plt.hist(V)
```

## Проверяем эффективность


```python
n=10
M=10000
X =np.random.normal(0,1, (M, n))
V = X.mean(axis = 1)
m = np.median(X, axis = 1)
plt.hist(V)
plt.hist(m, color = (0.1, 0.1,1, 0.4))
```

```python
n=10
M=10000
s = 30
X = np.insert(X, 3,values = s*np.ones(M),axis = 1)
V = X.mean(axis = 1)
m = np.median(X, axis = 1)
plt.hist(V)
plt.hist(m, color = (0.1, 0.1,1, 0.4))
```
