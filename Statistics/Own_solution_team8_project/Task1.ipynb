```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Task 1

```python
df =  pd.read_csv("df.csv",sep = ';', error_bad_lines = False)
del df['X\Y']
df = df.astype(float)
df
```

```python
X = df.sum(axis = 1)
X.name = 'X'
Y = df.sum(axis = 0)
Y.name = 'Y'
X.to_csv("vector_x.csv", header = True, index = False)
Y.to_csv("vector_y.csv", header = True, index = False)
print("Распределение X:\n", X.values)
print("\nРаспределение Y:\n", Y.values)
```

```python
sns.distplot(X)
sns.distplot(Y)
```
